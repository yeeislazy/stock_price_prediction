```python
import pandas as pd
from pathlib import Path
import numpy as np
```

```python
raw_data_path = Path.cwd().parent / "data" / "raw" / "aapl.parquet"

df = pd.read_parquet(raw_data_path)
df.head()
```

```python
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
```

```python
def compute_ma(df: pd.DataFrame, window: int=20) -> pd.DataFrame:
    df['ma' + str(window)] = df['close'].rolling(window=window).mean()
    return df
df = compute_ma(df)
df.iloc[-20:]
```

```python
def compute_rsi(df: pd.DataFrame, window=14) -> pd.DataFrame:
    delta = df['close'].diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=window).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=window).mean()
    rs = gain / loss
    rsi = 100 - (100 / (1 + rs))
    df['rsi' + str(window)] = rsi
    return df

df = compute_rsi(df)
df.iloc[-20:]
```

```python
def compute_atr(df: pd.DataFrame) -> pd.DataFrame:
    # compute ATR
    high_low = df['high'] - df['low']
    high_close = abs(df['high'] - df['close'].shift())
    low_close = abs(df['low'] - df['close'].shift())
    true_range = pd.DataFrame({'high_low': high_low, 'high_close': high_close, 'low_close': low_close}).max(axis=1)
    df['ATR'] = true_range.rolling(window=14).mean()
    return df
df = compute_atr(df)
df.iloc[-20:]
```

```python
def compute_pipeline(df: pd.DataFrame) -> pd.DataFrame:
    df = compute_ma(df)
    df = compute_rsi(df)
    df = compute_atr(df)
    df = df.dropna()
    return df
```
